```python
def DarRpp(R, Rp, E, E0, V, u):
    Rpp = -2/u*Rp + (-E/E0 + V)*R
    return Rpp
```

```python
z = 2
E0 = z**2 *(-13.6)
umax = 20.0
du = 10**(-4)
R0 = 1.0
Rp0 = -0.8
E =  25.422

u = arange(0.01, umax, du)
v = -1/u*(1+pow(e,-u))
print v
```

```python
plot (u,v)
```

```python
plot(log(u),log(-v),"k",label="Prueba")
plot(log(u),log(u**-1),"r--",label="Potencial cerca")
plot(log(u),log(2*u**-1),"b--",label="Potencial lejos")
legend()
xlabel('log u')
ylabel('log V')
savefig('log0.png')
```

```python
listR = zeros(len(u))
listRp  = zeros(len(u))
listRpp = zeros(len(u))

listR[0] = R0
listRp[0] = Rp0
listRpp[0] = DarRpp(listR[0], listRp[0], E, E0, v[0],u[0])

print listRpp
```

#intento #1

```python
n = 0
while n < len(u)-1:
    Utest = u[n]
    Rtest = listR[n]
    Rptest = listRp[n]
    Rpptest = listRpp[n]
    
    
    K1 = Rptest*du
    L1 = Rpptest*du
    
    Utest= u[n] + 0.5*du
    Rtest = listR[n]+ 0.5*K1
    Rptest = listRp[n] + 0.5*L1
    Rpptest = DarRpp(Rtest,Rptest,E, E0, v[n], Utest) 
    
    
    K2 = Rptest*du
    L2 = Rpptest * du
    
    Utest= u[n] + 0.5*du
    Rtest = listR[n]+ 0.5*K2
    Rptest = listRp[n] + 0.5*L2
    Rpptest = DarRpp(Rtest,Rptest,E, E0, v[n], Utest)
    
    K3 = Rptest*du
    L3 = Rpptest*du
    
    Utest= u[n] + 0.5*du
    Rtest = listR[n]+ 0.5*K3
    Rptest = listRp[n] + 0.5*L3
    Rpptest = DarRpp(Rtest,Rptest,E, E0, v[n], Utest)
    
    K4 = Rptest*du
    L4 = Rpptest*du
    
    K = (K4 + 2*K3 + 2*K2 + K1)/6.0
    L = (L4 + 2*L3 + 2*L2 + L1)/6.0
    
    listR[n+1] = listR[n] + K
    listRp[n+1] = listRp[n] + L
    listRpp[n+1] = DarRpp(listR[n+1], listRp[n+1], E, E0, v[n+1], u[n+1])
    
    n = n+1
```

```python
plot (u,listR)
title('0th Radial Function')
xlabel('u')
ylabel('R')
savefig('0R.png')
```

```python
listP = 4*pi* u**2 * listR**2
```

```python
const = 0
for i in listP:
    const = const + i
```

```python
listP = listP/const
```

```python
plot(u, listP)
title('0th Probability Density')
xlabel('u')
ylabel('P')
savefig('0P.png')
```

```python
listS = zeros(len(u))
for i in range(len(u)):
    listS[i] = listS[i-1] + 10**4*listP[i]*du
```

```python
plot (u, listS)
title('Cumulative Probability')
xlabel('u')
ylabel('S')
savefig('CumulativeP.png')
```

```python
Q = (2 - listS)
v = Q/(u**2)*du

plot (u,Q)
```
